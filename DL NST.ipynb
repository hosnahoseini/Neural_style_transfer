## Task 3 : Preprocess image

Torchvision models page : https://pytorch.org/docs/stable/torchvision/models.html

```python
from PIL import Image
from torchvision import transforms as T

def preprocess(img_path, max_size=500):
  image = Image.open(img_path).convert("RGB")

  if max(image.size) > max_size:
    size = max_size
  else:
    size = max(image.size)

  img_transforms = T.Compose([
      T.Resize(size),
      T.ToTensor(),
      T.Normalize(mean=[0.485, 0.456, 0.406],
                  std=[0.229, 0.224, 0.225])
   ])
  image = img_transforms(image)

  image = image.unsqueeze(0)

  return image
```

```python
from pygments import style
content_p = preprocess('/content/Project-NST/content11.jpg')
style_p = preprocess('/content/Project-NST/style12.jpg')

content_p = content_p.to(device)
style_p = style_p.to(device)

print("content shape:", content_p.shape)
print("style shape:", style_p.shape)
```

## Task 4 : Deprocess image

```python
import numpy as np
import matplotlib.pyplot as plt

def deprocess(tensor):

  image = tensor.to('cpu').clone()
  image = image.numpy()
  image = image.squeeze(0)
  image = image.transpose(1, 2, 0)
  image = image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
  image = image.clip(0, 1)

  return image
```

```python
content_d = deprocess(content_p)
style_d = deprocess(style_p)

print("deprocess content shape:", content_d.shape)
print("deprocess style shape:", style_d.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.imshow(content_d)
ax2.imshow(style_d)
```

## Task 5 : Get content,style features and create gram matrix

```python
def get_features(image, model):
  layers = {
      '0': 'conv1_1',
      '5': 'conv2_1',
      '10': 'conv3_1',
      '19': 'conv4_1',
      '21': 'conv4_2', #content_feature
      '28': 'conv5_1',
  }

  x = image
  Features = {}

  for name, layer in model._modules.items():

    x = layer(x)
    if name in layers:
      Features[layers[name]] = x

  return Features
```

```python
content_f = get_features(content_p, vgg)
style_f = get_features(style_p, vgg)
```

```python
def gram_matrix(tensor):
  b, c, h, w = tensor.size()
  tensor = tensor.view(c, h*w)
  gram = torch.mm(tensor, tensor.t())
  return gram
```

```python
styles_gram = {layer: gram_matrix(style_f[layer]) for layer in style_f}
```

## Task 6 : Creating Style and Content loss function

```python
def content_loss(target_conv_4_2, content_conv4_2):
  loss = torch.mean((target_conv_4_2 - content_conv4_2)**2)
  return loss
```

```python
style_weights = {
    'conv1_1': 1.0,
    'conv2_1': 0.75,
    'conv3_1': 0.2,
    'conv4_1': 0.2,
    'conv5_1': 0.2,

}
```

```python
def style_loss(style_weights, target_features, styles_gram):
  loss = 0

  for layer in style_weights:
    target_f = target_features[layer]
    target_gram = gram_matrix(target_f)
    style_gram = styles_gram[layer]
    b, c, h, w = target_f.shape
    layer_loss = style_weights[layer] * torch.mean((target_gram - style_gram) ** 2)
    loss += layer_loss/(c*h*w)

  return loss
```

```python
target = content_p.clone().requires_grad_(True).to(device)
target_f = get_features(target, vgg)
print("content loss: ", content_loss(target_f['conv4_2'], content_f['conv4_2']))
print("style loss: ", style_loss(style_weights, target_f, styles_gram))
```

## Task 7 : Training loop

```python
from torch import optim

optimizer = optim.Adam([target], lr=0.003)

alpha = 1
beta = 1e5

epochs = 3000
show_every = 500
```

```python
def total_loss(c_loss, s_loss, alpha, beta):
  loss = alpha * c_loss + beta * s_loss
  return loss
```

```python
result = []

for i in range(epochs):

  target_f = get_features(target, vgg)
  c_loss = content_loss(target_f['conv4_2'], content_f['conv4_2'])
  s_loss = style_loss(style_weights, target_f, styles_gram)
  t_loss = total_loss(c_loss, s_loss, alpha, beta)

  optimizer.zero_grad()
  t_loss.backward()
  optimizer.step()

  if i % show_every == 0:
    print(f"Total Loss at Epoch {i}: {t_loss}")
    result.append(deprocess(target.detach()))
```

```python
plt.figure(figsize=(10, 8))

for i in range(len(result)):
  plt.subplot(3, 2, i+1)
  plt.imshow(result[i])

plt.show()
```
